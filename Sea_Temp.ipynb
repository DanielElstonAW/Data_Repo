```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy import transpose as T
```

```python
min_temp = np.array([7.9,7.8,7.4,8.5,10.1,12,14.3,15.6,16.4,15.1,12.7,10.2])
max_temp = np.array([11.4,10.4,10,11.2,12.7,14.8,16.2,17.9,18.1,17.3,15.9,13.4])
dfw = pd.DataFrame([min_temp, max_temp])
dfw = dfw.T
dfw = dfw.rename(columns={0:'Min', 1:'Max'})
dfw
```

```python
dfw['Mean'] = dfw.Min + ((dfw.Max-dfw.Min)/2)
dfw
```

```python
dfw['Std'] = dfw.std(axis=1)
```

```python
dfw['n_days'] = [31,28,31,30,31,30,31,31,30,31,30,31]
dfw
```

```python
31*16*12
```

```python
dfw['cells_per_day'] = dfw['n_days'].apply(lambda x: x*12)
dfw
```

```python
t_obs = [np.random.normal(i,j,k) for i,j,k in zip(dfw['Mean'],dfw['Std'], dfw['cells_per_day'])]
```

```python
t_obs_rav = np.ravel(t_obs).tolist()
```

- Here we have created an array of normal distribution samples for sea temperatures.
- Min and Max values of sea temperatures are taken per month.
- A normal distribution has been created for each month.
- A random sample is extracted 31*16 times (days in a month * 4 * observations of engines per day).
- This creates the correct number of temperature observations for 1 years.
- We have 12 years and 3 months of observations. (12.25 years)
- Therefore we repeat this operation 12.25 times
