```python
ship_id = pd.DataFrame({'Ship_id': [i for i in range(1, n_ships+1)]})
ship_id = ship_id.loc[ship_id.index.repeat(16)]
ship_id = np.tile(np.array([ship_id.Ship_id]), 4473)
ship_id = pd.DataFrame(ship_id.reshape(-1,1), columns=['Ship_id'])
ship_id
```

```python
print(len(ship_id)/4)

eng_n = ['F1', 'F2', 'F3', 'F4']
eng_n_repeat = eng_n*(214704)
eng_n = pd.DataFrame(eng_n_repeat, columns=['eng_n'])
eng_n
```

---

```python
date_range = pd.date_range(start='1/1/2010', periods=4473, freq='D')
df_dr = pd.DataFrame(date_range, columns=['date'])
df_dr
```

```python
 insert_rows = 192

 df_dr.index = range(0, insert_rows * len(df_dr), insert_rows)
 date_range = df_dr.reindex(index = range(insert_rows * len(df_dr))) 
```

```python
date_range = date_range.ffill()
date_range
```

---

```python
n1 = ['HMS Richmond']
n2 = ['HMS Argyll']
n3 = ['HMS Iron Duke']
n4 = ['HMS Kent']
n5 = ['HMS Lancaster']
n6 = ['HMS Montrose']

n7 = ['HMS Northumberland']
n8 = ['HMS Portland']
n9 = ['HMS Somerset']
n10 = ['HMS St Albans']
n11 = ['HMS Sutherland']
n12 = ['HMS Wesminster']

name_list = [n1,n2,n3,n4,n5,n6,n7,n8,n9,n10,n11,n12]
```

```python
name_data = [i*16 for i in name_list]
name_data = np.ravel(name_data).tolist()
df_name = pd.DataFrame(name_data, columns=['Ship_Name'])
df_name = np.tile(np.array([df_name.Ship_Name]), 4473)
df_name = pd.DataFrame(df_name.reshape(-1,1), columns=['Ship_Name'])

df_name
```

---

```python
time_range = pd.date_range(start='1/1/2010', periods=4473*24*2, freq='0.25H').strftime('%H:%M:%S').repeat(4)
time_range
```

```python
time_range = pd.DataFrame(time_range, columns=['time'])
time_range
```

---

```python
frames = [ship_id.reset_index(drop=True), df_name.reset_index(drop=True), eng_n, time_range, date_range]
  
df = pd.concat(frames, axis=1)
df
```
